####Bigram Probability

```python
files = open("Largest_chapter_2.txt",encoding='latin-1')
wordslists = files.read().splitlines() # to escape \n occurence
wordslists = [i for i in wordslists if i!='']
w= ""
w = w.join(wordslists)
```

```python
words=w.split()
bigrams = [(words[i], words[i + 1]) for i in range(len(words) - 1)]
from collections import Counter

bigram_counts = Counter(bigrams)
bigram_probabilities = {}
for bigram, count in bigram_counts.items():
    word1, word2 = bigram
    bigram_probabilities[bigram] = count / words.count(word1)

from tabulate import tabulate
table_data = []
for bigram, probability in bigram_probabilities.items():
    word1, word2 = bigram
    table_data.append([word1, word2, probability])


headers = ['Word 1', 'Word 2', 'Probability']


print(tabulate(table_data, headers=headers, tablefmt='grid'))
```

```python
def make_guess(previous_word, bigram_probabilities):
    next_word_probabilities = {word: prob for (prev, word),prob in bigram_probabilities.items() if prev == previous_word}
    sorted_guesses = sorted(next_word_probabilities, key=next_word_probabilities.get, reverse=True)
    return sorted_guesses
```
